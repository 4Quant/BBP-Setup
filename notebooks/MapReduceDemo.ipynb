```
L = ["cat dog car", "dog car dog"]
sentence_rdd = sc.parallelize(L)
print(sentence_rdd)
```

```
word_rdd = sentence_rdd.flatMap(lambda sentence: [(word, 1) for word in sentence.split(" ")])
print(word_rdd, word_rdd.first(), 'Count', word_rdd.count())
```

```
word_count_rdd = word_rdd.reduceByKey(lambda a_count, b_count: a_count+b_count)
print(word_count_rdd.collect())
```

```
wc_df = word_count_rdd.map(lambda x: {'word': x[0], 'count': x[1]}).toDF()
```

```
wc_df.createOrReplaceTempView("WordCount")
```

```
sqlContext.sql("""
SELECT * FROM WordCount
""").collect()
```
